```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("medical_examination.csv")
df.head()
```

```python
height_Meters = (df['height']/100)*(df['height']/100)
h_m = df['weight'] / height_Meters 
df.loc[h_m <= 25, 'overweight'] = 0 
df.loc[h_m > 25, 'overweight'] = 1 
df.head()
```

```python
df.loc[df['cholesterol']==1, 'cholesterol'] = 0 
df.loc[df['cholesterol']>1, 'cholesterol'] = 1 
df.loc[df['gluc']==1, 'gluc'] = 0 
df.loc[df['gluc']>1, 'gluc'] = 1 
df
```

```python
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['active','alco','cholesterol','gluc','overweight','smoke'])
df_cat
```

```python
sns.catplot(x='variable',col='cardio',hue='value',kind='count',data=df_cat)
```

```python
df_cat = df_cat.groupby(['cardio', 'variable', 'value']).size().reset_index()
df_cat = df_cat.rename(columns={0: 'total'})
df_cat
```

```python
graph = sns.catplot(data=df_cat, kind="bar", x="variable", y="total", hue="value", col="cardio")
fig = graph.fig
```

```python
df_heat = df[(df['ap_lo'] <= df['ap_hi']) &
                 (df['height'] >= df['height'].quantile(0.025)) &
                 (df['height'] <= df['height'].quantile(0.975)) &
                 (df['weight'] >= df['weight'].quantile(0.025)) &
                 (df['weight'] <= df['weight'].quantile(0.975))
                 ]
df_heat
```

```python
corr = df_heat.corr()
corr
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
mask
```

```python
 fig, ax = plt.subplots(figsize=(16, 9))
 fig, ax
```

```python
sns.heatmap(corr, mask=mask, square=True, linewidths=0.5, annot=True, fmt="0.1f")
```
